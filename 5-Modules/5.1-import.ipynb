##### Importing Modules in Python:Modules and packages 


In python,modules and packages help organize and reuse code. 

```python
import math
math.sqrt(16)
```

```python
from math import sqrt,pi
print(sqrt(16))
print(sqrt(25))
print(pi)
```

```python
import numpy as np
np.array([1,2,3,4,5])
```

```python
from math import *

print(sqrt(16))
pi
```

```python
from package.maths import addition
addition(2,3)
```

```python
## another way
from package import maths
maths.addition(2,3)
```
