```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
dataset=pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header=1)
dataset.head()
```

```python
dataset.info()
```

### Data Cleaning


```python
## missing values
dataset[dataset.isnull().any(axis=1)]
```

The dataset is converted into two sets based on region from 122th index,we can make a coulmn based on the region 

1: "Bejaia Region Dataset"

2: "Sidi-Bel Abbes Region Dataset"

Add new column with region

```python
dataset.loc[:122,"Region"]=0
dataset.loc[122:,"Region"]=1
df=dataset
```

```python
df.info()
```

```python
df[['Region']]=df[['Region']].astype('int64')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
## Remove null values
df=df.dropna().reset_index(drop=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
## Need to remove the above as it is repeated 
df=df.drop(index=122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
## fix spaces in columns
df.columns=df.columns.str.strip()
df.columns
```

### Change the required data into integer

```python
df[['day','month','year','Temperature','RH','Ws']]=df[['day','month','year','Temperature','RH','Ws']].astype('int')
```

```python
df.info()
```

```python
objects=[features for features in df.columns if df[features].dtypes=='O']
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python

df.to_csv('algerian_forest_fires_cleaned.csv',index=False)
```

#### EDA


```python
## drop day, month, year
df_copy=df.drop(['day','month','year'],axis=1)
```

```python
df_copy.head()
```

```python
## Encoding the categories in classes
df_copy['Classes']=df_copy['Classes'].astype(str)
df_copy['Classes']= np.where(df_copy['Classes'].str.contains('not fire',case=False,na=False),0,1)
```

```python
## plot density for all features
plt.style.use('seaborn-v0_8')
df_copy.hist(bins=30,figsize=(20,15))
plt.show()
```

```python
## percentage of piechart
percentage=df_copy['Classes'].value_counts(normalize=True)*100
```

```python
## plotting piechart
labels=['not fire','fire']
plt.figure(figsize=(12,7))
plt.pie(percentage,labels=labels,autopct='%1.1f%%',startangle=30 )
plt.title('Percentage of fire and not fire cases')
plt.show()
```
