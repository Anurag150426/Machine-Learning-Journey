##### File Operation- Read And Write Files

File Handeling is crucial part of any proggraming language. Python provides built-in functions and methods to read and write to files,both binary and text.

```python
### Read a whole file

with open('example.txt','r')as file:
    content= file.read()
    print(content)
```

```python
## Read a file line by line

with open('example.txt','r') as file:
    for line in file:
        print(line.strip())## strip() removes newline charcter
```

```python
## Writing a file(Overwriting)

with open('example.txt','w')as file:
    file.write('Hello world!\n')
    file.write('this is a new line.')
```

```python
## Writing a file(without Overwriting)
with open('example.txt','a') as file:
    file.write("append operation taking place/n")
```

```python
## writing a list of lines to a file
lines=['first line \n','second line \n','third line \n']
with open('example.txt','a')as file:
    file.writelines(lines)
```

```python
### Read the content from a source text file and write to a destination text file
# copying to text file

with open('example.txt','r') as source_file:
    content=source_file.read()

with open('destination.txt','w') as destination_file:
    destination_file.write(content)    
```

The w+ mode in Python is used to open a file or both reading and writing.If file does not exists it is created.If the file exists,its content is truncated(i.e the file is overwrritten)

```python
### Writing and then reading a file

with open('example.txt','w+') as file:
    file.write("hello world\n")
    file.write("This is a new line")

    ## Move the file cursor to the begining
    file.seek(0)

    ## Read the content of the file
    content=file.read()
    print(content)
```
