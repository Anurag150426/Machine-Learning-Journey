```python
## why functiobs
num=24
if num%2==0:
    print("the number is even")
else:
    print("number is odd")    
```

```python
def even_or_odd(num):
    """This functions find even or odd"""
    if num%2==0:
        print("the number is even")
    else:
        print("the number is odd")    
```

```python
## call this function
even_or_odd(24)
```

```python
## function with multiple parametrs
def add(a,b):
    """This function adds two numbers"""
    c=a+b
    return c

result=add(2,4)
print(result)
```
