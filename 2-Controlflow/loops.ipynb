### While Loop


```python
## while loop

## The while loop continues to execute as long as condition is true

count=0

while count<5:
    print(count)
    count=count+1
```

```python
count=0

while count%2==0:
    print(count)
    count=count+1
```

```python
## Loop control statements
## break

for i in range(10):
    if i==5:
     break
    print(i)
```

```python
## continue

## The continue statement skips the current iteration and continues with the next.

for i in range(10):
    if i%2==0:
        continue
    print(i)


```

```python
## pass
## The pass statement is a null operation; it does nothing.

for i in range(5):
    if i==3:
        pass
    print(i)
```

```python
## Nested loopss
## a loop inside a loop

for i in range(3):
    for j in range(2):
        print(f"i:{i} and j:{j}")
```

```python
## Nested loopss
## a loop inside a loop
rows=5
for i in range(rows):
    for j in range(i+1):
        print("*",end=" ")
    print(" ")
```

```python
## Calculate the sum of first N numbers using while loop
## While loop
n=10
sum=0
count=1
while count<=n:
    sum=sum+count
    count=count+1
print("The sum of first n numbers are:",sum)
```
