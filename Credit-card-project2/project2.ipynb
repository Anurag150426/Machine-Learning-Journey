#### Question 1
 Write a query to print top 5 cities with highest spends and their percentage contribution of total credit card spends


```python
top_cities= df.groupby('City')['Amount'].sum()
top_cities.sort_values(ascending=False).reset_index()
top_cities.head().reset_index()
```

```python
## percentage
total_amount=df['Amount'].sum()
print(total_amount)
```

```python
percentage= (top_cities/total_amount)%100
print(percentage)
```

```python
top5_cities= pd.DataFrame({
    'amount': top_cities ,'percentage_of_total':percentage
}).reset_index()
answer_1=top5_cities.head()
print(answer_1)
```

#### Question 2
 Write a query to print highest spend month and amount spent in that month for each card type.


```python
df['Y-M']=df['Date'].dt.to_period('M')

monthly_spend= df.groupby(['Y-M','Card Type'])['Amount'].sum().reset_index()
monthly_spend.head(5)

```

```python
highest_month=monthly_spend.sort_values(['Card Type','Amount'],ascending=False)
card_Spend=highest_month.groupby('Card Type').first().reset_index()
card_Spend.head()
```

#### Question_3

```python
df_sorted_value=df.sort_values(by=['Card Type','Date'])
df_sorted_value['Cumulative_spend']=df_sorted_value.groupby('Card Type')['Amount'].cumsum()

df_greater= df_sorted_value[df_sorted_value['Cumulative_spend']>=1000000]

results=df_greater.groupby('Card Type').first().reset_index()
results.head()
```

#### question_7

```python
df['Y-M']=df['Date'].dt.to_period('M')
```

```python
month= df.groupby(['Card Type', 'Exp Type', 'Y-M'])['Amount'].sum().reset_index()
```

```python
month['Month']=month['Y-M'].dt.to_timestamp()
```

```python
month['prev_month']=month.groupby(['Card Type','Exp Type'])['Amount'].shift(1)
```

```python
month['growth']=(month['Amount'] - month['prev_month']) / month['prev_month'] * 100
```

```python
growth_jan=month[month['Month'].dt.strftime('%Y-%m')=='2014-01']
```

```python
result=growth_jan.sort_values('growth', ascending=False)
```

```python
result_7=result.head(1)
print(result_7)
```

#### question_8

```python
df['Day_of_week']=df['Date'].dt.dayofweek
```

```python
weekends= df[df['Day_of_week'].isin([5,6])]
```

```python
spend_count=weekends.groupby('City').agg(total_amount=('Amount','count'),Transaction=('Amount','count')).reset_index()
```

```python
spend_count['Spend_Transaction_ratio']=spend_count['total_amount']/spend_count['Transaction']
```

```python
ans=spend_count.sort_values('Spend_Transaction_ratio',ascending=False)
answer=ans.head(1)
print(ans.head(1))
```

#### question_9

```python
df=df.sort_values(by='Date')
```

```python
df['Trans_Count'] = df.groupby('City').cumcount() + 1

trans_500 = df[df['Trans_Count'] == 500]

first_trans = df.groupby('City')['Date'].min().reset_index(name='First_Date')

merged = pd.merge(trans_500[['City', 'Date']], first_trans, on='City')
merged['Days_Taken'] = (merged['Date'] - merged['First_Date']).dt.days

result_9 = merged.sort_values('Days_Taken')
answer_9=result_9.head(1)
print(answer_9)
```
