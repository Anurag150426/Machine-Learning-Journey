```python
## Series

## A pandas series is a 1-dimensional array-like object that can hold any datatype.It is similar to coulmn in a table

import pandas as pd
data=[1,2,3,4,5]
series=pd.Series(data)
print("Series \n",series)
```

```python
## create a series from dictinary
data={'a':1,'b':2,'c':3}## a,b,c becomes the index earlier it is 1,2,3 default
series_dict=pd.Series(data)
print(series_dict)
```

```python
data=[10,20,30]
index=['a','b','c']
pd.Series(data,index)
```

```python
## Dataframe 
## create a Dataframe from a dictionary of lists
data={
    'Name':['Anurag','Sanjay','Ani'],
    'Age':[21,55,37],
    'City':['Jaipur','Tezu','Sunpura']
}
df=pd.DataFrame(data)
print(df)
print(type(df))
```

```python
import numpy as np
np.array(df)
```
